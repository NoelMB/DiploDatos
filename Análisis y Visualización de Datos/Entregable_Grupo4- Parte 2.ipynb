## Ejercicio 1: Estimación

**Consigna:**  Calcular una estimación puntual y un intervalo de confianza de nivel (1-alpha) para la resta entre la media del salario Neto para Hombres menos la media del salario Neto para otros géneros(diferencia de las medias entre el grupoA y grupoB).
¿Cómo se relaciona este intervalo de confianza con el test de hipótesis?

```
# 1. Estimación puntual (diferencia de medias)
media_hombres = np.mean(groupA)
media_otros = np.mean(groupB)
estimacion_puntual = media_hombres - media_otros
print(f"Estimación puntual (media hombres - media otros): {estimacion_puntual:.2f}")

# 2. Intervalo de confianza
# Prueba de diferencia de medias (asumiendo varianzas distintas - Welch's t-test)
alpha = 0.05  # para un 95% de confianza
t_stat, p_value = stats.ttest_ind(groupA, groupB, equal_var=False)

# Grados de libertad aproximados
n1 = len(groupA)
n2 = len(groupB)
s1_sq = np.var(groupA, ddof=1)
s2_sq = np.var(groupB, ddof=1)

se = np.sqrt(s1_sq/n1 + s2_sq/n2)  # Error estándar de la diferencia
df = ( (s1_sq/n1 + s2_sq/n2)**2 ) / ( ((s1_sq/n1)**2)/(n1-1) + ((s2_sq/n2)**2)/(n2-1) )

# Valor crítico t
t_crit = stats.t.ppf(1 - alpha/2, df)

# Intervalo de confianza
ci_lower = estimacion_puntual - t_crit * se
ci_upper = estimacion_puntual + t_crit * se
print(f"Intervalo de confianza al 95%: ({ci_lower:.2f}, {ci_upper:.2f})")
```

Test de hipótesis: \\
Hipótesis Nula: $H_0$: $\mu_A-\mu_B=0$ \\
Hipótesis Alternativa: $H_1$: $\mu_A-\mu_B \neq 0$

Si el intervalo de confianza contiene el 0, entonces NO rechazamos $H_0$.

Si el intervalo de confianza NO contiene el 0, entonces rechazamos $H_0$ al nivel de significancia $α$.

En resumen, el intervalo de confianza y el test de hipótesis son dos formas diferentes de responder la misma pregunta sobre la diferencia de medias.

## Ejercicio 2: Test de hipótesis



### 2.1 Formalización

Describir formalmente los distintos componentes de un test de hipótesis para comprobar si la distribución de los salarios es distinta entre los grupos A y B.




Hipótesis Nula: $H_0$: $\mu_A-\mu_B=0$. No hay diferencia en las medias poblacionales de los salarios. \\
Hipótesis Alternativa: $H_1$: $\mu_A-\mu_B \neq 0$. Hay diferencia en las medias poblacionales de los salarios.

Tipo de test: bilateral

Estadístico de de prueba (pivote): $\frac{(\bar{X}_A-\bar{X}_B)-0}{\sqrt{\frac{s^2_A}{n_A}+\frac{s^2_B}{n_B}}}$

Distribución bajo $H_0$
Bajo la hipótesis nula $H_0$, el estadístico $T$ sigue aproximadamente una distribución t de Student con grados de libertad dados por la fórmula de Welch

### 2.2 P-valor

1. Calcule el p-valor y decida si rechazar o no la hipótesis nula.
2. Interprete el resultado.

Links útiles:
* [Test de hipótesis usando scipy](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.ttest_ind.html)
* [Test de Welch](http://daniellakens.blogspot.com/2015/01/always-use-welchs-t-test-instead-of.html)

```
# 6. Prueba de hipótesis (t-test de Welch)
t_stat, p_value = stats.ttest_ind(groupA, groupB, equal_var=False)
#print(f"Valor t: {t_stat:.2f}")
print(f"p-valor: {p_value:.20f}")
```

Como el $p-valor \leq \alpha$ entonces se rechaza $H_0$. \\
En general, cuanto menor es el valor de $p$, mayor es la evidencia que hay en los datos de la muestra en contra de la hipótesis nula y la hipótesis alternativa. Es decir, $H_0$ debe ser rechazada a favor de $H_1$, cuando el valor $p$ es suficientemente pequeño.

CONCLUSIÓN: Hay una diferencia significativa entre ambos grupos.
